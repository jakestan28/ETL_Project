```python
# Import SQL Alchemy
import sqlalchemy
from sqlalchemy import create_engine, inspect, func
# Import and establish Base for which classes will be constructed 
from sqlalchemy.ext.declarative import declarative_base
# from sqlalchemy.ext.automap import automap_base
Base = declarative_base()
# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float
```

```python
class Players(Base):
    __tablename__ = 'stats'
    id = Column(Integer, primary_key=True, index=True)
    club_id = Column(Float)
    first = Column(String(30))
    last = Column(String(30))
    nation = Column(String(30))
    position = Column(String(30))
    club = Column(String(30))
    age = Column(Float)
    born = Column(Float)
    mp = Column(Float)
    starts = Column(Float)
    minutes = Column(Float)
    goals = Column(Float)
    assists = Column(Float)
    pk = Column(Float)
    pkatt = Column(Float)
    yellow_card = Column(Float)
    red_card = Column(Float)
```

```python
# Create Database Connection
# ----------------------------------
# Establish Connection
engine = create_engine('postgresql+psycopg2://postgres:postgres@localhost/mls_db')
conn = engine.connect()
```

```python
# Create the team names table within the database
Base.metadata.create_all(conn)
```

```python
# clear out the db
Base.metadata.drop_all(engine)
```

```python
# Create a Session Object to Connect to DB
# ----------------------------------
# Session is a temporary binding to our DB
from sqlalchemy.orm import Session
session = Session(bind=engine)
```

```python
for player in second_df.iterrows(): 
#     print(player[1][0], player[1][1],player[1][2],player[1][3],player[1][4],player[1][5],player[1][6],player[1][7],player[1][8],player[1][9],player[1][10],player[1][11],player[1][12],player[1][13],player[1][14], player[1][15],
#     player[1][16],player[1][17])
    data = Players(club_id=player[1][0],first=player[1][1],last=player[1][2],nation=player[1][3],position=player[1][4],club=player[1][5],age=player[1][6],born=player[1][7],mp=player[1][8],starts=player[1][9],
    minutes=player[1][10],goals=player[1][11],assists=player[1][12],pk=player[1][13],pkatt=player[1][14],yellow_card=player[1][15],red_card=player[1][16])
    session.add(data)
    session.commit()
```
